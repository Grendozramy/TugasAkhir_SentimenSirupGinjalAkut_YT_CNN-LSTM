```python
import pandas as pd
import nltk

# Labeling
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Import dataset
df = pd.read_csv('../../Data/2. Clean/clean copy.csv')

# Menampilkan total dataset
print("Total Dataset :", len(df))

df.head()
```

```python
# Membuat kolom baru untuk menyimpan label sentimen dan skor sentimen
df['sentiment_label'] = ''
df['sentiment_score'] = ''

# Inisialisasi objek SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Fungsi untuk melakukan analisis sentimen pada setiap komentar
def sentiment_analysis(comment):
    # Tambahkan pengecekan apakah nilai comment sudah diubah menjadi string
    if isinstance(comment, float):
        return '0'
    # Menghitung skor sentimen pada komentar
    scores = sid.polarity_scores(comment)
    # Menentukan label sentimen berdasarkan skor sentimen
    if scores['compound'] >= 0.05:
        return '1'
    elif scores['compound'] <= -0.05:
        return '-1'
    else:
        return '0'

# Looping untuk melakukan sentiment_label kata dan analisis sentimen pada setiap komentar
for i, row in df.iterrows():
    comment = row['text_manual_replaced']
    # Tambahkan pengecekan apakah nilai comment sudah diubah menjadi string
    if isinstance(comment, str):
        sentiment_label = sentiment_analysis(comment)
        sentiment_score = sid.polarity_scores(comment)['compound']
        df.at[i, 'sentiment_label'] = sentiment_label
        df.at[i, 'sentiment_score'] = sentiment_score
    else:
        df.at[i, 'sentiment_label'] = '0'
        df.at[i, 'sentiment_score'] = 0.0
```
