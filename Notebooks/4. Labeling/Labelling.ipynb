```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import dataset
df = pd.read_csv('../../Data/3. EDA/clean-eda-label.csv')

# Menampilkan total dataset
print("Total Dataset :", len(df))

df.head()
```

```python
# Mengecek isi kolom khusus
df['label'].unique()
```

```python
# Menghitung jumlah kemunculan setiap nilai pada kolom khusus
df['label'].value_counts()
```

Distribusi data (Label)

```python
# Count the values in the label column
label_counts = df['label'].value_counts()

# Plot the distribution of the label column
plt.bar(label_counts.index, label_counts.values, tick_label=['-1', '0', '1'])
plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Distribution of Label')
plt.show()
```

Boxplot

```python
# Group the data by the 'label' column
grouped_data = df.groupby('label')

# Create an empty list to store the data for each label
data_to_plot = []

# Iterate through the groups and append the data to the list
for name, group in grouped_data:
    data_to_plot.append(group['label'])

# Create a boxplot of the data for each label
plt.boxplot(data_to_plot)
plt.ylabel('Label')
plt.title('Boxplot of Label')
plt.xticks([1, 2, 3], ['-1', '0', '1'])  # Set custom x-axis tick labels
plt.show()
```
