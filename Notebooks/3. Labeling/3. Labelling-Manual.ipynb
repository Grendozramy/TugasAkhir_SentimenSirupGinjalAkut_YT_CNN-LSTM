```python
# Import dataset
df = pd.read_csv('/content/drive/Othercomputers/My Laptop/Telkom University/Perkuliahan/S8 TA/Source Code/Data/3. EDA/clean-eda.csv')

# Menampilkan total dataset
print("Total Dataset :", len(df))

df.head()
```

```python
# Membuat kolom baru untuk menyimpan labeling manual
df['manual_labels'] = ''

# Fungsi untuk melakukan labeling manual pada setiap komentar
def label_sentiment(comment):
    print(comment)
    label = int(input("Masukkan label sentimen (1 = positif, 0 = netral, -1 = negatif): "))
    return label

# Looping untuk melakukan labeling kata dari labeling manual pada setiap komentar
for i, row in df.iterrows():
    comment = row['raw']
    label = label_sentiment(comment)
    df.at[i, 'manual_labels'] = label
```
